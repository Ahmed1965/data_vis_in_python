```python
import seaborn as sns
import matplotlib.pyplot as plt
phool = sns.load_dataset('iris')
#print(iris)
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.xlim(4)
plt.ylim(2)
plt.title("Phool")
plt.show()
```
